```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import StandardScaler
y = df['price_range']
X = df.drop('price_range',axis=1)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,random_state=42,shuffle=False)
```

```python
scalar = StandardScaler()
X_train = scalar.fit_transform(X_train)
X_test = scalar.transform(X_test)
```

# Decision Tree 

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
parms = {'criterion':['gini', 'entropy', 'log_loss'],
         'splitter':['best', 'random'],
         'max_depth':[1,5,10,15,50,100],
         'min_samples_split' :[1,3,5,15,50,100],
         'min_samples_leaf':[1,2,5,15,50,100],
         'max_features':['auto','sqrt','log2'],
         'random_state':[0,1,41,42]}
gcv = GridSearchCV(estimator=model,param_grid=parms,cv= 5)
gcv.fit(X_train,y_train)
```

```python
gcv.best_estimator_
```

```python
gcv.best_score_
```

```python
best_model = gcv.best_estimator_
```

```python
best_model.fit(X_train,y_train)
test_pred = best_model.predict(X_test)
train_pred = best_model.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

test_score = accuracy_score(y_test,test_pred)
train_score = accuracy_score(y_train,train_pred)
```

```python
print('Test_Score:',test_score)
print('train_score:',train_score)

print('Train_classification_Report',classification_report(y_train,train_pred))
print('Test_classification_Report',classification_report(y_test,test_pred))
```

the differance between the train and the test accuracy is more which means that the model is over fitted 

```python
train_CM = confusion_matrix(y_train,train_pred)
test_CM = confusion_matrix(y_test,test_pred)
```
